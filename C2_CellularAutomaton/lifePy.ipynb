```python
import random as rnd
import pyglet

class GameOfLife:

    def __init__(self, window_width,windo_height , cell_size):
        self.grind_width  = int(window_width / cell_size) #Tamaño de la malla 
        self.grind_height = int(windo_height / cell_size)
        self.cell_size    = cell_size #Tamaño celula
        self.cells        = []
        self.generate_cells()


    def generate_cells(self): #Genera la matrz
        for row in range(0, self.grind_height):
            self.cells.append([])
            for col in range(0, self.grind_width):
                if rnd.random() < 0.4 : 
                    self.cells[row].append(1)
                else:
                    self.cells[row].append(0)

    def run_rules(self): #Reglas del mismo ejercicio que se hizo en clase ... el que se hizo a mano 
        temp = []
        for row in range(0, self.grind_height):
            temp.append([])
            for col in range(0, self.grind_width):  
                cell_sum = sum([self.get_cell_value(row - 1,col    ),
                                self.get_cell_value(row - 1,col - 1),
                                self.get_cell_value(row    ,col - 1),
                                self.get_cell_value(row + 1,col - 1),
                                self.get_cell_value(row + 1,col    ),
                                self.get_cell_value(row + 1,col + 1),
                                self.get_cell_value(row    ,col + 1),
                                self.get_cell_value(row - 1,col + 1)])

                if self.cells[row][col] == 0 and cell_sum==3:
                    temp[row].append(1)
                elif self.cells[row][col] == 1 and (cell_sum == 3 or cell_sum ==2):
                    temp[row].append(1)
                else: 
                    temp[row].append(0)

        self.cells = temp
    
    def get_cell_value(self,row,col):
        if row >= 0 and row<self.grind_height and col >= 0 and col < self.grind_width: #Obtiene el valor de cada celula
            return self.cells[row][col]
        return 0 



    def draw(self): #Dibuja la AC en pantalla
        for row in range(0, self.grind_height):
            for col in range(0, self.grind_width):  
                if self.cells[row][col] == 1:
                    #[si row = col = 0] (0,0)(0,20)(20,0)(20,20)
                    square_coords = (row * self.cell_size , col * self.cell_size,
                                    row * self.cell_size , col * self.cell_size + self.cell_size,
                                    row * self.cell_size + self.cell_size , col * self.cell_size,
                                    row * self.cell_size + self.cell_size , col * self.cell_size + self.cell_size)
                    pyglet.graphics.draw_indexed(4,pyglet.gl.GL_TRIANGLES,
                                                [0,1,2,1,2,3],
                                                ('v2i',(square_coords)))
```

```python
import pyglet
from gameoflive import GameOfLife #Modulo creado tambien adjunto

class Window(pyglet.window.Window):

    def __init__(self):
        super().__init__(500,500)
        self.gameoflife = GameOfLife(self.get_size()[0], #ancho
                                    self.get_size()[1],  #alto
                                     5) #tamaño de la celula
        pyglet.clock.schedule_interval(self.update,1.0/20.0) #tiempo en que tarda en cambiar de estado 

    def on_draw(self):
        self.clear()
        self.gameoflife.draw() #Dibuja las celulas en pantalla
        

    def update(self, dt):
        self.gameoflife.run_rules() #Reglas del automata celular

if __name__ =='__main__':
    window = Window()
    pyglet.app.run() #Corre el programa
```
