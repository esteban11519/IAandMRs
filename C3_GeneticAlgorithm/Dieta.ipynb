Para solucionar este problema, se crean clases tipo alimento, dieta y posible dieta, donde con esto es posible almacenar toda la información relacionada a propiedades alimenticias de alimentos, sus cantidades correspondientes con base a la decodificación de su cromosoma asignado. También se tiene la dieta deseada como una clase, en este caso es una dieta de  2000 calorías, 250 g de carbohidratos, 75 g de proteinas y 65 g de grasas.



```python
from numpy.random import randint
from numpy.random import rand
import random 
import math
```

```python
class Alimento:
    def __init__(self, nombre, calorias,carbohidratos,proteinas,grasas):
        self.nombre = nombre
        self.calorias=calorias
        self.carbohidratos=carbohidratos
        self.proteinas=proteinas
        self.grasas=grasas
        self.cromosoma=0
        self.cantidad=0


class Dieta:
    def __init__(self,calorias,carbohidratos,proteinas,grasas):
        self.calorias=calorias
        self.carbohidratos=carbohidratos
        self.proteinas=proteinas
        self.grasas=grasas

class PosibleDieta:
    def __init__(self,alimentos):
        self.alimentos=alimentos
```

```python
def genera(K,l,A):
    N=l*A
    pob = [randint(0, 2, N).tolist() for _ in range(K)]
    return pob
```


Para la función aptitud, se intenta estimar lo mejor posible a la dieta, sin embargo, si se obtiene un valor para alguno de estos 4 parámetros mayor a su 10%, se considera unválida la solución. La aptitud global de cada posible dieta surge del promedio geométrico de la aptitud proveniente de los 4 parámetros de la dieta.

```python
def eval_apt(pob,l,A,TablaAlimentos,dieta):
    apt_crom = []
    apt_pob = 0
    for i in range(0, len(pob)):
        crom = pob[i]
        apt = Func_ap(crom,l,A,TablaAlimentos,dieta)
        apt_pob += apt
        apt_crom.append(apt)
    probab = []
    for i in range(0, len(pob)):
        prob_crom = apt_crom[i]/apt_pob
        probab.append(prob_crom)
    return probab

def Func_ap(cromosoma,l,A,TablaAlimentos,dieta):
    n=[]
    for i in range (A):
        n.append(decodificar(cromosoma[(l*i):(l*(i+1))],l))

    return evaluarDieta(dieta,n,TablaAlimentos)


def evaluarDieta(dieta,n,TablaAlimentos):

    Totales=TotalesDieta(n,TablaAlimentos)
    Tcalorias=Totales.get("calorias")
    Tcarbohidratos=Totales.get("carbohidratos")
    Tproteinas=Totales.get("proteinas")
    Tgrasas=Totales.get("grasas")

    Icalorias=dieta.calorias
    Icarbohidratos=dieta.carbohidratos
    Iproteinas=dieta.proteinas
    Igrasas=dieta.grasas

    Acalorias=Aptitud(Icalorias,Tcalorias)
    Acarbohidratos=Aptitud(Icarbohidratos,Tcarbohidratos)
    Aproteinas=Aptitud(Iproteinas,Tproteinas)
    Agrasas=Aptitud(Igrasas,Tgrasas)

    AptitudTotal=(Acalorias*Acalorias*Acarbohidratos*Aproteinas*Agrasas)**(1/5)
    return AptitudTotal


def TotalesDieta(n,TablaAlimentos):
    Tcalorias=0
    Tcarbohidratos=0
    Tproteinas=0
    Tgrasas=0

    for i in range(len(n)):
        Tcalorias+=TablaAlimentos[i].calorias*n[i]/100
        Tcarbohidratos+=TablaAlimentos[i].carbohidratos*n[i]/100
        Tproteinas+=TablaAlimentos[i].proteinas*n[i]/100
        Tgrasas+=TablaAlimentos[i].grasas*n[i]/100
    
    totales={"calorias":Tcalorias,
            "carbohidratos":Tcarbohidratos,
            "proteinas":Tproteinas,
            "grasas":Tgrasas}

    return totales

def Aptitud(I,T):
    if(T>1.1*I):
        return 0
    else:
        return 1-abs(I-T)/I


def imprimirDieta(cromosoma,TablaAlimentos,A):

    n=[]
    for i in range (A):
        n.append(decodificar(cromosoma[(l*i):(l*(i+1))],l))

    Totales=TotalesDieta(n,TablaAlimentos)

    Tcalorias=Totales.get("calorias")
    Tcarbohidratos=Totales.get("carbohidratos")
    Tproteinas=Totales.get("proteinas")
    Tgrasas=Totales.get("grasas")
    print("\nTotales:")
    print("Calorias:"+str(int(Tcalorias))+"g")
    print("Carbohidratos:"+str(int(Tcarbohidratos))+"g")
    print("Proteinas:"+str(int(Tproteinas))+"g")
    print("Grasas:"+str(int(Tgrasas))+"g")

    print("\nCantidad de alimentos: ")
    for i in range (len(TablaAlimentos)):
        if(n[i]>0):
            print(TablaAlimentos[i].nombre+": "+str(n[i])+"g")


def decodificar(cromosoma,l): 
    xi = 0  # Limite inferior en el dominio real
    xf = 100  # Limite superior en el dominio real
    Max = 2 **(l)
    cromPot = [cromosoma[i]*2**(l-i-1) for i in range(l)]
    valorDecimal = sum(cromPot) 
    cromPot = 0
    valDeco = int((((xf-xi)/Max)*(valorDecimal)+xi)/5)*5
    return valDeco
```

```python
def seleccion(pob, probab):
    j = 0
    K = len(pob)
    limite = 2 * max(probab)
    pob_nueva = []
    while j < K: 
        i = 0 
        while i < K: 
            aleat = random.uniform(0, limite)
            if probab[i] > aleat: 
                pob_nueva.append(pob[j])
                j += 1
                if j >= K:
                    break
            i +=1
            if i == K:
                i = 0
    return pob_nueva
```

```python
def cruce(pob_nueva): 
    i = 0 
    K = len(pob_nueva)
    hijos =[]
    while i < K:
        crom1 = pob_nueva[i]
        crom2 = pob_nueva[i+1]
        pt = randint(1, len(crom1)-1)
        hijo1 = crom1[:pt] + crom2[pt:]
        hijo2 = crom2[:pt] + crom1[pt:]
        hijos.append(hijo1)
        hijos.append(hijo2)
        i = i +2; 
    return(hijos)
```

```python
def mutacion(hijos, pm, l):
    K = len(hijos)
    totalbits = K * l 
    segmento = 1/pm 
    n_mutaciones = totalbits/segmento
    i = 0 
    while i < n_mutaciones: 
        muta = randint(0, totalbits-1)
        x = math.floor(muta/l) 
        y = muta%l
        if hijos[x][y] == 0: 
            hijos[x][y] = 1 
        else: 
            hijos[x][y] = 0 
        i += 1
    return hijos
```

```python
def Alg_Genetico(l, M, K, p_mut,A,TablaAlimentos,dieta): 
    pob = genera(K,l,A)  #Se genera aleatoriamente la población 
    probab = eval_apt(pob,l,A,TablaAlimentos,dieta) #Se evalúa esa primera población
    mejor_apt_AG = 0
    mejor_cromo_AG = [0 for _ in range(l*A)]
    i =0;                    #Es la primera generación 
    while i < M:     #Mientras no se ejecuten todas las generaciones
        n_pob = seleccion(pob, probab) #Selecciona, los nuevo padres  
        hijos = cruce(n_pob)           #Cruce de los padres generan hijos
        pob = mutacion(hijos, p_mut, l)   #Se mutan los hijos para generar otra población       
        probab = eval_apt(pob,l,A,TablaAlimentos,dieta) #Se evalua la población i, se calcula la probabilidad de selección           
        mejor_prob = max(probab)
        mejor_posic = probab.index(max(probab))
        mejor_crom = pob[mejor_posic]
        mejor_apt = Func_ap(mejor_crom,l,A,TablaAlimentos,dieta)
        if mejor_apt > mejor_apt_AG:
            mejor_apt_AG = mejor_apt
            mejor_cromo_AG = mejor_crom
        

        if(mejor_apt>0.98):break

        i = i + 1  

    print("Generación "+str(i)+": Aptitud: "+str(mejor_apt))
    imprimirDieta(mejor_crom,TablaAlimentos,A)

                             #Se incrementa el contador de generaciones
    #print("generacion ", i, "Mejor solución  x", xAG, "valor de la función", yAG)
```
